### Structure your Notebook 

#### Don't leave an 'Untitled' Title
Anyone should be able to name a notebook something helpful, so that if the notebook is passed to any second person it should be easily understandable what for the notebook is.
Follow a naming convention that shows the owner and the order the analysis was done in. We will use the format 
```bash
<step>-<ghuser>-<description>.ipynb
```
(e.g., <b>0.3-user1-visualize-distributions.ipynb</b>)

#### Give a proper title

Give your notebook a title (H1 header) and a meaningful preamble to describe its purpose and contents.

#### Headings and Documentation
Use headings and documentation in Markdown cells to structure your notebook and explain your workflow steps. Remember: You’re not only doing this for your colleagues or your successor, but also for your future self.

#### Add table of content (toc2)

By enabaling <b>toc2</b> jupyter extension, you can automatically create heading numbers and a Table of Contents, both in a sidebar (optionally a floating window) and in a markdown cell. The highlighting indicates your current position in the document — this will help you keep oriented in long notebooks.

### Refactor & outsource code into modules

#### PEP8 Standard

Stick to the standards of good coding — think Clean Code principles and PEP8. Use meaningful variable and function names, comment sensibly, modularize your code and don’t be too lazy to refactor.

#### Modularization
After you’ve written plain code in cells to get ahead quickly, acquire the habit of <b>turning stable code into functions and move them to a dedicated module.</b> This makes your notebook more readable and is incredibly helpful when productionizing your workflow. Like below:

This:

    df = pd.read_csv(filename)
    df.drop( ...
    df.query( ...
    df.groupby( ...
    
becomes this:

    def load_and_preprocess_data(filename):
       """DOCSTRING"""
       # do stuff
       # ...
       return df

and finally this:

    import dataprep
    df = dataprep.load_and_preprocess_data(filename)

#### Autoreload

If you edit a module file, Jupyter’s autoreload extension reloads imported modules:

    %load_ext autoreload
    %autoreload 2
    
Other useful magic syntax can be found here: https://towardsdatascience.com/jupyter-notebook-hints-1f26b08429ad

### Embrace reproducibility

#### Version Control
Learn to use git efficiently and make a habbit of pushing and commiting your code for a single change you make in
your notebook workflow

#### 3.2 Remove Outputs before pushing

Depending on your project and purpose, it might be reasonable to use a <b>git pre-commit hook that removes notebook output</b>, like nbstripout. It will make commits and diffs more readable, but might discard output (plots etc.) you actually want to store. Also it prevents client data leackage chances from third party software. 

#### Same environment for same notebook

Run your notebook in a dedicated (conda/docker) environment. <b>Store the requirements.txt file in the git repository alongside your notebooks and modules</b>. This will help you reproduce your workflow as well as facilitate transitioning into a production environment.

### Be familiar with Productive Jupyter Extensions

The <b>jupyter_contrib_nbextensions</b> package contains a collection of community-contributed unofficial extensions that add functionality to the Jupyter notebook. These extensions are mostly written in Javascript and will be loaded locally in your browser.

#### How to install
To install the jupyter_contrib_nbextensions notebook extensions, three steps are required.   

* First, the Python pip package needs to be installed. 
* Then, the notebook extensions themselves need to be copied to the Jupyter data directory. 
* Finally, the installed notebook extensions can be enabled, either by using built-in Jupyter commands, or more conveniently by using the

##### pip install
```bash
$ pip install jupyter_contrib_nbextensions
```

##### Install javascript and css files
```bash
$ jupyter contrib nbextension install --user
```

##### Scratchpad

Adds a scratchpad cell to Jupyter notebook. This is a cell in which you can execute code against the current kernel without modifying the notebook document.

The scratchpad can be toggled by clicking the icon in the bottom-right.
![alt text](https://jupyter-contrib-nbextensions.readthedocs.io/en/latest/_images/demo3.gif "Title Text 1")

##### autopep8

Autopep8 helps to reformat/prettify the contents of code cells with just a click. If you are tired of hitting the spacebar again and again to format the code, autopep8 is your savior.
![alt text](https://miro.medium.com/max/1026/1*hO8tQavKgqiyj0ZKFOVSPg.gif "Title Text 1")

##### qgrid

Qgrid is a Jupyter notebook widget which uses SlickGrid to render pandas DataFrames within a Jupyter notebook. This allows you to explore your DataFrames with intuitive scrolling, sorting, and filtering controls, as well as edit your DataFrames by double clicking cells.

Installation guide:

```bash
pip install qgrid
jupyter nbextension enable --py --sys-prefix qgrid
# only required if you have not enabled the ipywidgets nbextension yet
jupyter nbextension enable --py --sys-prefix widgetsnbextension
```
![alt text](https://miro.medium.com/max/1400/1*xGvLi1RhdIdSbCA9cPsS2A.gif "Title Text 1")

##### Hinterland

It enables code autocompletion menu for every keypress in a code cell, instead of only calling it with the tab. This makes Jupyter notebook’s autocompletion behave like other popular IDEs such as PyCharm.
![alt text](https://miro.medium.com/max/1104/1*pOEq37eIbKH3AwKgyfDB2w.gif "Title Text 1")

##### Snippets

This extension adds a drop-down menu to the Notebook toolbar that allows easy insertion of code snippet cells into the current notebook.
![alt text](https://miro.medium.com/max/1400/1*uoxIUbG6m6YiJbjVu77rGQ.gif "Title Text 1")

#### Executing Shell Commands

Any command that works at the command-line can be used in IPython by prefixing it with the <b>!</b> character.

```bash
In [1]: !ls
example.jpeg list tmp

In [2]: !pwd
/home/Parul/Desktop/Hello World Folder

In [4]: files= !ls

In [5]: print(files)
['example.jpeg', 'list', 'tmp']
```
